# About Data
### This dataset features Netflix content—both `TV shows` and `movies` released between `2008` and `2021`.

- type: Indicates whether the entry is a “Movie” or a “TV Show”.
- title: The official title of the movie or TV show.
- director: Name(s) of the director(s). May be blank if unknown.
- cast: Main actors or actresses, listed as comma-separated names.
- release_year: Year when the movie or show was released (2008–2021).
- rating: The audience rating category (e.g., PG-13, TV-Mature, etc.).
- duration: For movies, formatted as “X min”; for TV shows, specified as “Y Seasons”.
- listed_in: Genre(s) of the content, comma-separated (e.g., Drama, Sci-Fi & Fantasy).
- description: A brief synopsis of the movie or show.

**Handling Nulls**

```python
# Replace Null values

df["director"].fillna("No Director", inplace=True)
df["cast"].fillna("No Cast", inplace=True)
df["country"].fillna("No Country", inplace=True)
df.dropna(subset=["date_added", "rating", "duration"], inplace=True)
```

```python
df.isnull().any()
```

---
## 🔧**Feature Engineering**

- extracte the day and month from the 'date_added' column to make it easier to analyze trends over time.
- Create a new column for genre to extract insights.
- Convert the 'date_added' column to datetime format.
- Drop unnecessary columns

```python
# Convert 'date_added' to datetime format with 'coerce' to handle and replace invalid dates with NaT.
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')

# Extracted month,month_name,day  information from the 'date_added' column
df['month_name'] = df['date_added'].dt.month_name()
df['day'] = df['date_added'].dt.day_name()
```

```python
df.head(3)
```

```python
#Create a new column for genre
genres = df['listed_in'].str.split(', ')

df['genre1'] = genres.str[0]
df['genre2'] = genres.str[1]  
df['genre3'] = genres.str[2]  
```

## 📺**Movies vs. Shows**

```python
df["type"].value_counts()
```

```python
def my_fmt(x):
    return '{:.2f}%\n({:.0f})'.format(x, total*x/100)

values = df.type
v_counts = values.value_counts()
total = len(values)

plt.Figure(figsize=(14,6))
plt.pie(v_counts, explode=(0.025, 0.025), labels=v_counts.index, autopct=my_fmt, startangle=180, shadow=True, colors=["Gray", "Red"]);
plt.title('Movies vs TV Shows')
plt.show()
```

#### 💡The number of `movies` on Netflix is more than double the number of `TV shows`.This highlights Netflix’s stronger focus on film content over episodic series.
---

## 🎬 **Netflix Genre Preferences Analysis**

```python
# Get top 10 genres for movies and shows
tv_genre = df[df['type']=='TV Show']['genre1'].value_counts().head(10)
movie_genre = df[df['type']=='Movie']['genre1'].value_counts().head(10)

fig, axes = plt.subplots(1, 2, figsize=(14,6))
sns.barplot(x=tv_genre.values, y=tv_genre.index, palette="mako", ax=axes[0])
axes[0].set_title("Top 10  Genres (TV Shows)")

sns.barplot(x=movie_genre.values, y=movie_genre.index, palette="mako", ax=axes[1])
axes[1].set_title("Top 10 Genre (Movies)")

plt.tight_layout()
plt.show()
```

#### 💡`International TV Shows` dominate the genre preferences among the top TV shows on Netflix, while `Drama` stands out as the primary choice for movies
---

## 🗳️ **Distribution of Viewers Ratings**

```python
# ratings = df.rating.value_counts().index[0:11]

tv_ratings = df[df["type"] == "TV Show"]["rating"].value_counts().head(7)
movie_rating = df[df["type"] == "Movie"]["rating"].value_counts().head(7)

fig, axes = plt.subplots(1, 2, figsize=(14,6))
sns.barplot(x=tv_ratings.values, y=tv_ratings.index, palette="mako", ax=axes[0])
axes[0].set_title("Top Ratings (TV Shows)")

sns.barplot(x=movie_rating.values, y=movie_rating.index, palette="mako", ax=axes[1])
axes[1].set_title("Top Ratings (Movies)")

plt.tight_layout()
plt.show()
```

#### 💡 `TV-MA` is the most common viewer rating for movies as well as TV Shows, indicating a strong presence of content suitable for mature audiences.

## 📈**Releases Over the Year**

```python
df.release_year.value_counts().to_frame('Value_count').head(20)
```

```python
df2 = df[df.release_year > 2009][["type", "release_year"]]
movie_year = df2[df2["type"] == "Movie"].release_year.value_counts().to_frame().reset_index()
show_year = df2[df2["type"] == "TV Show"].release_year.value_counts().to_frame().reset_index()

fig, ax = plt.subplots(figsize=(14, 6))
sns.lineplot(data=movie_year, x="release_year", y="count", marker="o", color="Black")
sns.lineplot(data=show_year, x="release_year", y="count", marker="o", color="Red")

plt.title("Total Releases Each Year")
plt.legend(["Movie", "TV Show"])
plt.xlabel("Years")
plt.ylabel("Count")
plt.show()
```

#### 💡 `Movie` releases on Netflix peaked in `2017–2018` with over `750` titles. `TV Show` releases steadily grew until `2020`.
